Below the python coding ,I used basic symbols to calculate minutes into total seconds for 5 seasons using multiplication (*) and addition (+).The most important thing was how to use floor division (//) and the Modulator (%). ### Find the highest Score

In Week 2, I learned structural logic in Python by working with numerical ranges, loops, and conditional statements (if, elif, else). The below coding is about to check and find the highest and second highest score.The program loops through a user's defined number of student scores, continuously and comparing each input against the current maximum values. By using conditional checks, score will be updates and stores the top two distinct values into highest_score and second highest variables.After all the step done it will show the output for highest score and second highest score

```python
#ask use to input number of students
num_student = int(input("Enter number of students: "))
#list to store scores
highest_score = 0
second_highest = 0
#find the highest and second_highest score
for i in range(num_student):
    score=int(input(f"Enter student {i+1} score: "))
#check conditions
    if score > highest_score:
        second_highest = highest_score   
        highest_score = score            

    elif score > second_highest:
        second_highest = score
#out put for calculation results
print(f'The highest score is {highest_score}')
print(f'The second highest score is {second_highest}')
```

### Password cracker

I also learned how to create a password guesser coding that ask the user to input string data entry in second coding as shown in below. After that analyze the user input in len() function to keep track of the string limit.And use Boolean to check True and False variables to handle tracking indicators. Using min() function, the code makes sure that the correct range is obtained for matching purposes without raising an index error. If the user enters the right password in the first attempt, then the code exits successfully.If not the code will check the length and correction data that the user input and it will shows the accurancy

```python
#create password
#ask to guess the password
password = "MDX123"
guess  = input("Guess the Password : ")
#check guess is correct or not
if guess == password:
    print("Password cracked!")
else:
    correct = 0                         #to calculate matches num or letter
    used    = [False] * len(password)   #prevent to calculate false twice
#first loop to check for correct position and letter
    for i in range(min(len(password), len(guess))):
        if guess[i] == password[i]:
            correct += 1  
            used[i] = True

#second loop to check for correct letter but wrong position
    for i in range(len(guess)):
        for j in range(len(password)):
            if (not used[j] and guess[i] == password[j]):
                correct += 1
                used[j] = True
                break
#cacculte accuracy
    accuracy = (correct / len(password)) * 100
    print(f"accurancy {accuracy:.0f} %")
```

In week 4, I learned about data cleaning and making lists using built-in functions and methods such as split(), strip(), count(), and len(). In this program, I used split() and strip() to create lists and delete unnecessary spaces and characters. Specifically, split() breaks a single string into a list of smaller strings based on a designated separator, while strip() removes the unwanted characters from the edges. Because these are string methods, you must place a dot (.) before them. Additionally, len() is a built-in function that counts the number of elements or characters; you pass the target object inside its parentheses. Finally, although count() is a method that can be used directly on strings, lists, or tuples, this specific program calculates how many 'ERROR', 'INFO', and 'WARNING' words are included in the data by using a for loop and counter variables.

```python
#create a list of logs
logs = [
    "[14:32:05] ERROR: Disk full",
    "[14:32:10] INFO: Server started",
    "[14:33:01] WARNING: High memory usage",
    "[14:33:15] INFO: Backup complete",
    "[14:34:00] ERROR: Connection refused",
    "[14:35:22] WARNING: CPU at 90%"
]
#use def function to classify logs
def classify_logs(logs):
    errors_count = 0
    infos_count = 0
    warning_count = 0

# Extract the timestamp (between the square brackets)
    for log in logs:
        parts=log.split("]")
        timestemp=parts[0].strip("[")
        message_count=parts[1].strip()
        timestemp=parts[0].strip("[")

# Print each log with its timestamp
        print(f"timestemp {timestemp}")

# Increment the appropriate counter
# Check if the text after the timestamp starts with ERROR, INFO, or WARNING
        if message_count.startswith("ERROR"):
            errors_count+=1
        elif message_count.startswith("INFO"):
            infos_count+=1
        elif message_count.startswith("WARNING"):
            warning_count+=1
    return errors_count,infos_count,warning_count

errors,infos,warning=classify_logs(logs)
#for the output
print(f"ERROR {errors}")
print(f"INFOS {infos}")
print(f'WWARNING {warning}')
    
    
```

## Password Strength Check

The second program is a password strength checker that determines if a test password is weak, medium, or strong. First, I created a test_password list and defined a function using def. If a password contains any spaces, the program prints an error stating that spaces are not allowed and stops checking.If not a score variable is created to track the password's strength. The program adds 1  to the score for each condition met, if the length is 8 or more characters, and if it contains uppercase letters, lowercase letters, digits, or special characters. I used if ,elif and elase statements to classify the total score. A score is equal to 2 or lower than 2 ,it will shows the password is weak, a score is  equal to 4 or lower than 4 ,it will shows medium, and any other score will shows strong.

```python
#create a list of passwords
test_passwords = [
    "hello",
    "Hello123",
    "My Pass!",
    "Str0ng@Pass",
    "12345678"
]
#use def function to check password
def check_password(password):

# password deny if space include
    if " " in password:
        print(f"Password: '{password}' Invalid password: Spaces are not allowed.")
        return 

    score = 0

# At least 8 characters
    if len(password) >= 8:
        score += 1

# Contains uppercase letter
    if any(char.isupper() for char in password):
        score += 1

# Contains lowercase letter
    if any(char.islower() for char in password):
        score += 1

 # Contains a digit
    if any(char.isdigit() for char in password):
        score += 1

# check Contains a special character (@#$%!)
    special_characters = "@#$%!"
    if any(char in special_characters for char in password):
        score += 1

    if score <= 2:
        label = "Weak"
    elif score <= 4:
        label = "Medium"
    else:
        label = "Strong"

#For the output    
    print(f"Password: '{password}'")
    print(f"Score: {score}/5 - {label}")



# check password 
for pw in test_passwords:
    check_password(pw)
```

### Managing Study Group

In week 5, I learned about file handling and data structures, how to read and write data to and from text and CSV files. In this program, I created study_group dictionary container to store student names as keys and sets of their study subjects as values. To add the new "Biology" group for John, I used the  set method .add(). To make a student leaving a group, "Emily" as student_to_leave and "Physics" as former_group.And then  I used if statement to safely check the conditions before using .remove() method to eliminate the former group from her set. I used with open  in write mode "w"  with f.write() to save the updated data directly into study_group.txt.After that to search 'John' with update data ,use if ,else to check conditions

```python
#create study group
study_group = {
    "John": {"Math", "Physics"},
    "Emily": {"Physics", "Biology"},
    "Alex": {"Math"}
}

#reate student to add new group
student_join_group = "John"
new_group = "Biology"
if student_join_group not in study_group:
    study_group[student_join_group] = set()
study_group[student_join_group].add(new_group)

# Create leave student
student_to_leave = "Emily"
former_group = "Physics"

if student_to_leave in study_group and former_group in study_group[student_to_leave]:
    study_group[student_to_leave].remove(former_group)

#Save to file 
with open("study_group.txt", "w") as f:
    for student, group in study_group.items():
        group_string = ",".join(group)
        f.write(f"{student}:{group_string}\n")

# Search student
search_student = "John"
if search_student in study_group and study_group[search_student]:
    display_text = ", ".join(study_group[search_student])
    print(f"{search_student} is part of these groups: {display_text}")
else:
    print(f"{search_student} is not in any study groups.")
```

## Counting words in Les_Brown



In this program, I import re (Regular Expressions) library into my script to perform text cleaning on complex data. First, I  create a dictionary named count_list to store the final data. I used a with opne  in read mode 'r' with read_lines() to read the file line by line. strip() removes whitespace, and lower() converts every word to lowercase to make sure not to wrong with upper case ,it will change every charactor to lowercase .Then I used split to break each line down into individual words. Inside the loop, re sub() strips away punctuation characters from each word. If the cleaned word is valid, it is tracked inside the dictionary. I used  again with open() in write mode 'w' with f.write() to save the updated word counts to a new file named word_count.txt.

```python
#import re library for word counting
import re
count_list = {}

with open("week05_data/DATA-20260527/Les_Brown.txt", "r") as f:
    lines = f.readlines()
# Process each line to count words
for line in lines:
    clean = line.strip().lower()    #use .lower () to make every words to be lowercase
    if not clean:
        continue
        
    words = clean.split()
    for i in words:
        word = re.sub(r"[^\w]", "", i)
        if not word:
            continue
        if word not in count_list:
            count_list[word] = 1
        else:
            count_list[word] += 1
            
#write new data using write to the file word_count.txt
with open("week05_data/DATA-20260527/word_count.txt", "w") as f:
    for word, count in count_list.items():
        f.write(f"{word}: {count}\n")
```
